# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np

normal_vector = np.random.normal(size = 100)
print(f"Normal Vector: \n {normal_vector}")

chi_square_vector = np.random.chisquare(df = 3, size = 100)
print(f"Vector chisquare: \n {chi_square_vector}")

````

> #### Measures of Central Tendency

##### Mean (Average)

````python
import statistics as stats

# mean of normal_vector
print(f"Normal Vector mean: {stats.mean(normal_vector):.4f}")

# mean of chi-square distribution
print(f"Chi-square distribution mean: {stats.mean(chi_square_vector):.4f}")
````

##### Median (Middle Value)

````python
# Median of normal vector
print(f"Median of normal vector: {stats.median(normal_vector):.4f}")
# Median of chi-square disctribution
print(f"Media of chi-square distribution: {stats.median(chi_square_vector):.4f}")
````

##### Mode (Most Frequent Value)

````python
# Mode of normal vector
print(f"Mode of Normal vector: {stats.mode(normal_vector):.4f}")
# Mode of chi-square distribution
print(f"Mode of chi-square distribution: {stats.mode(chi_square_vector):.4f}")
````

> #### Measures of Dispersion

#### Range (Difference between max value and min value)

````python
print(f"Range of normal vector: {max(normal_vector) - min(normal_vector)}")
print(f"Range of chi-square distribution: {max(chi_square_vector) - min(chi_square_vector)}")
````

#### Variance and Standard deviation

````python
# variance and Standard deviation (normal_vector)
print(f"Variance of normal vector: {stats.variance(normal_vector):.4f}")
print(f"Standard deviation of normal vector: {stats.stdev(normal_vector):.4f}")

# variance and Standard deviation (chi-square)
print(f"\nVariance of chi-square: {stats.variance(chi_square_vector):.4f}")
print(f"Standard deviation of chi-square: {stats.stdev(chi_square_vector):.4f}")
````

> #### Shape Measures

#### Skewness

````python
import matplotlib.pyplot as plt
from scipy.stats import skew, norm

def plot_skewness(vector_data):
  plt.hist(vector_data, bins = 30, alpha = 0.5, color = 'g', density = True)
  xmin, xmax = plt.xlim()
  x = np.linspace(xmin, xmax, 100)
  p = norm.pdf(x, stats.mean(vector_data), stats.stdev(vector_data))
  plt.plot(x, p, 'k', linewidth = 2)
  title = 'Skewness = %.2f' % (skew(vector_data))
  plt.title(title)
  plt.show()

# Normal vector
plot_skewness(normal_vector)

# chi-square
plot_skewness(chi_square_vector)
````

#### Kurtosis

````python
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, norm

def plot_kurtosis(vector_data):
  plt.hist(vector_data, bins = 30, alpha = 0.5, color = 'r', density = True)
  xmin, xmax = plt.xlim()
  x = np.linspace(xmin, xmax, 100)
  p = norm.pdf(x, stats.mean(vector_data), stats.stdev(vector_data))
  plt.plot(x, p, 'k', linewidth = 2)
  title = 'Kurtosis = %.2f' % (kurtosis(vector_data))
  plt.title(title)
  plt.show()

# Normal vector
plot_kurtosis(normal_vector)

# chi-square distribution
plot_kurtosis(chi_square_vector)
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import statistics

def avg_calc(data):
  if len(data) <= 1:
    return data[0]
  else:
    return sum(data) / len(data)

def sd_calc(data):
  if len(data) <= 1:
    return 0.0
  else:
    return statistics.stdev(data)
  
data = [4, 2, 5, 8, 6]
print(f"Data: {data}")
print(f"Standard deviation: {sd_calc(data)}")
````
